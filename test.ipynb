```python
from datasets import get_dataset_config_names, load_dataset

# Get all available config names for the ai4bharat/MILU dataset
configs = get_dataset_config_names("ai4bharat/MILU", trust_remote_code=True)

# Select the Hindi config
hindi_config = None
for config in configs:
    if "hindi" in config.lower():
        hindi_config = config
        break

target_domains = [
    'Arts & Humanities',
    'Law & Governance',
    'Social Sciences',
    'Health & Medicine'
]

if hindi_config:
    print(f"\nSelected Hindi config: {hindi_config}")
    dataset = load_dataset("ai4bharat/MILU", hindi_config, trust_remote_code=True)
    # For each domain, print stats per split
    domain_total = 0
    for domain in target_domains:
        print(f"\nDomain: {domain}")
        split_domain_total = 0
        for split in dataset.keys():
            ds_split = dataset[split]
            if "domain" in ds_split.column_names:
                mask = [d == domain for d in ds_split["domain"]]
                count = sum(mask)
                print(f"  Split '{split}': {count} examples")
                split_domain_total += count
        print(f"  Total in '{domain}': {split_domain_total} examples")
        domain_total += split_domain_total
    print(f"\nTotal number of examples in all selected domains: {domain_total}")
else:
    print("No Hindi config found in ai4bharat/MILU.")
```

```python
from datasets import load_dataset
import ast
from collections import Counter

# List of target subjects (Indonesian-specific)
selected_subjects = [
    "Indonesian language",
    "Civic education",
    "History",
    "Islam religion",
    "Christian religion",
    "Hindu religion",
    "Geography",
    "Sociology",
    "Minangkabau culture",
    "Social science",
    "Art"
]

config = "default"
dataset = load_dataset("indolem/IndoMMLU", config, trust_remote_code=True)
print(f"Loaded dataset for config: {config}")

def string_to_list(s):
    try:
        return ast.literal_eval(s)
    except Exception:
        return []

for split in dataset.keys():
    data = dataset[split]
    print(f"\n=== '{split}' split: Option list lengths per Indonesian subject ===")
    global_length_counts = Counter()
    for subj in selected_subjects:
        subject_examples = [ex for ex in data if ex.get("subject") == subj]
        length_counts = Counter()
        for ex in subject_examples:
            options = ex.get("options", [])
            if isinstance(options, str):
                options = string_to_list(options)
            if isinstance(options, list):
                option_len = len(options)
                length_counts[option_len] += 1
                global_length_counts[option_len] += 1
        if subject_examples:
            print(f"Subject: {subj}")
            for num_options, count in sorted(length_counts.items()):
                print(f"  {count} example(s) with {num_options} options")
    # Also print total number of examples for selected subjects (optional)
    total_examples = sum(1 for ex in data if ex.get("subject") in selected_subjects)
    print(f"Total number of examples for selected subjects: {total_examples}")
    if total_examples == 0:
        print(f"  No matching subjects found in split '{split}'")
    # Print total number for each option length across all selected subjects
    if global_length_counts:
        print("Option count totals across all selected subjects in this split:")
        for num_options, count in sorted(global_length_counts.items()):
            print(f"  {count} example(s) with {num_options} options")
```

```python
from datasets import load_dataset

# Target configs of interest
target_configs = [
    "Turkish_Language_and_Literature",  # 土耳其语言与文学
    "History",                          # 历史
    "Religion_and_Ethics",              # 宗教与伦理
    "Geography",                        # 地理
]

dataset_id = "AYueksel/TurkishMMLU"

grand_total = 0
for config in target_configs:
    ds = load_dataset(dataset_id, config, trust_remote_code=True)
    num_test = len(ds["test"]) if "test" in ds else 0
    num_dev = len(ds["dev"]) if "dev" in ds else 0
    total = num_test + num_dev
    print(f"Config: {config}")
    print(f"  test split: {num_test} samples")
    print(f"  dev split:  {num_dev} samples")
    print(f"  total:      {total} samples\n")
    grand_total += total

print(f"Grand total number of examples across selected configs: {grand_total}")
```

```python
from datasets import load_dataset, get_dataset_config_names

dataset_id = "yangzhang33/culture-eval-benchmark-labeled"
configs = get_dataset_config_names(dataset_id)
print(f"Available configs for {dataset_id}: {configs}")

for config in configs:
    print(f"\nLoading config: {config}")
    ds = load_dataset(dataset_id, config)
    total_claude_cs = 0
    total_samples = 0
    for split in ds:
        split_data = ds[split]
        num_samples = len(split_data)
        total_samples += num_samples
        # Filter examples with claude_cs==1
        if "claude_cs" in split_data.column_names:
            count = sum(1 for v in split_data["claude_cs"] if v == 1)
            print(f"  {split}: {count} samples with claude_cs==1 ({num_samples} total samples)")
            total_claude_cs += count
        else:
            print(f"  {split}: claude_cs column not found ({num_samples} total samples)")
    print(f"Total samples with claude_cs==1 in {config}: {total_claude_cs}")
    print(f"Total number of samples in {config}: {total_samples}\n")
```
